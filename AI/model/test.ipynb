```python
import pandas as pd
import joblib
import numpy as np

model = joblib.load("crop_model.pkl")
encoder = joblib.load("label_encoder.pkl")

sample = pd.DataFrame({
    "N": [20],
    "P": [30],
    "K": [60],
    "temperature": [20.8],
    "humidity": [82.0],
    "ph": [6.5],
    "rainfall": [202.9]
})

# Probabilitas setiap kelas
prob = model.predict_proba(sample)[0]

# Urutkan dari probabilitas terbesar
top3 = np.argsort(prob)[::-1][:3]

print("="*60)
print("TOP 3 RECOMMENDATION")
print("="*60)

for rank, idx in enumerate(top3, start=1):
    crop = encoder.inverse_transform([idx])[0]
    print(f"{rank}. {crop:<15} {prob[idx]*100:.2f}%")
```
